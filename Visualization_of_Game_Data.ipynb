```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
battle = pd.read_csv('battles.csv')
```

```python
battle.head(6)
```

```python
battle.tail(6)
```

```python
battle.shape
```

```python
battle.rename(columns={'attacker_1':'main_attacker'},inplace=True)
battle.rename(columns={'defender_1':'main_defender'},inplace=True)
battle.head()
```

```python
battle['attacker_king'].value_counts()
```

```python
battle['location'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(13,8)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(16,8)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'])
plt.show()
```

```python
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
plt.show()
```
