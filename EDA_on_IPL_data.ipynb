```python
#loading the required libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading the ipl matches dataset
ipl=pd.read_csv('matches.csv')
```

```python
#having a glance at the first five records of the dataset
ipl.head()
```

```python
#having a glance at the last five records of the dataset
ipl.tail()
```

```python
ipl.shape
```

```python
#Getting the frequency of most man of the match awards
ipl['player_of_match'].value_counts()
```

```python
#Getting the top 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:10]
```

```python
#Getting the last 10 players with most man of the match awards
ipl['player_of_match'].value_counts()[-10:]
```

```python
#Getting the top 5 players with most man of the match awards
ipl['player_of_match'].value_counts()[0:5]
```

```python
#making a bar-plot for the top 5 players with most man of the match awards
plt.figure(figsize=(15,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()),list(ipl['player_of_match'].value_counts()[0:10]),color="maroon")
plt.show()
```

```python
#Getting the frequency of result column
ipl['result'].value_counts()
```

```python
#Finding out the number of toss wins w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
#Looking at the head
batting_first.head()
```

```python
#Making a histogram
plt.figure(figsize=(10,10))
plt.hist(batting_first['win_by_runs'])
plt.xlabel("Range of Runs")
plt.grid(True)
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first
batting_first['winner'].value_counts()
```

```python
#Making a bar-plot for top 7 teams with most wins after batting first
plt.figure(figsize=(15,7))
plt.bar(list(batting_first['winner'].value_counts()[0:7].keys()),list(batting_first['winner'].value_counts()[0:7]),color=["blue","yellow","silver","purple","red","orange","pink"])
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#extracting those records where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
#looking at the head
batting_second.head()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],color="palegreen")
plt.show()
```

```python
#Making a histogram for frequency of wins w.r.t number of wickets
plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'],bins=30,color="violet")
plt.show()
```

```python
#Finding out the frequency of number of wins w.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
#Making a bar plot for top-7 teams with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:7].keys()),list(batting_second['winner'].value_counts()[0:7]),color=["cyan"])
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.show()
```

```python
#Making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(15,12))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.2f%%')
plt.show()
```
